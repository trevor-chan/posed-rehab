```python
'''['neck', 0
 'nose', 1
 'l_sho', 2
 'l_elb', 3
 'l_wri', 4
 'l_hip', 5
 'l_knee', 6
 'l_ank', 7
 'r_sho', 8
 'r_elb', 9
 'r_wri', 10
 'r_hip', 11
 'r_knee', 12
 'r_ank', 13
 'r_eye', 14
 'l_eye',  15
 'r_ear', 16
 'l_ear'] 17'''

relevant_pairs = [[0,1],[0,2],[2,3],[3,4],[0,8],[8,9],[9,10],[0,5],[5,6],[6,7],[0,11],[11,12],[12,13],[1,14],[1,15],[14,16],[15,17]]
relevant_triplets = [[1,0,2],[1,0,8],[2,0,8],[0,2,3],[0,8,9],[2,3,4],[8,9,10],[0,5,6],[0,11,12],[5,6,7],[11,12,13]]
triplet_labels = ['left-neck', 'right-neck', 'left-right-shoulder', 'left-shoulder', 'right-shoulder', 'left-elbow', 'right-elbow', 'left-hip', 'right-hip', 'left-knee', 'right-knee']
```

```python
def center_pose(pose):
    pose = pose.reshape(-1,4)-np.tile(pose[0][8:12], (19,1))
    pose = np.delete(pose, 2, axis=0)
    return pose

def gaussian_smooth_positions(pose_timeline):
    return np.expand_dims(np.transpose(np.array(
        [gaussian_filter1d(np.array(pose_timeline)[:,0,i], 1) for i in range(np.array(pose_timeline).shape[2])]), (1,0)), axis=1)

def calc_angle(a, b, c):
    # a, b, c are three numpy arrays with x,y,z coordinates
    ba = a - b
    bc = c - b
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(cosine_angle)
    return np.degrees(angle)

def calc_joint_pos(pose):
    assert len(pose)==18
    angles = []
    for tri in relevant_triplets:
        angles.append(calc_angle(pose[tri[0]], pose[tri[1]], pose[tri[2]]))
    return np.array(angles)


lr_diff = ['shoulder', 'elbow', 'hip', 'knee']

def calc_left_right_diff(joint_config):
    assert len(joint_config)==11
    pairs = [[3,4],[5,6],[7,8],[9,10]]
    diffs = []
    for pair in pairs:
        diffs.append(np.abs(joint_config[pair[0]]-joint_config[pair[1]]))
    return diffs

def load_pose_sequence(filepath):
    sequence = pickle.load(open( filepath, "rb" ) )
    sequence = [pose[1] for pose in sequence]
    sequence = gaussian_smooth_positions(sequence)
    return [center_pose(pose) for pose in sequence]
```

```python
import cv2
import numpy as np

MAX_FEATURES = 200
GOOD_MATCH_PERCENT = 0.15

def alignImages(im1, im2):
    # Convert images to grayscale
    im1Gray = im1
    im2Gray = im2

    # Detect ORB features and compute descriptors.
    orb = cv2.ORB_create(MAX_FEATURES,fastThreshold=1, edgeThreshold=1)
    keypoints1, descriptors1 = orb.detectAndCompute(im1Gray, None)
    keypoints2, descriptors2 = orb.detectAndCompute(im2Gray, None)
    
        
    # Match features.
    matcher = cv2.DescriptorMatcher_create(cv2.DESCRIPTOR_MATCHER_BRUTEFORCE_HAMMING)
    matches = matcher.match(descriptors1, descriptors2, None)
    # Sort matches by score
    matches.sort(key=lambda x: x.distance, reverse=False)
    # Remove not so good matches
    numGoodMatches = int(len(matches) * GOOD_MATCH_PERCENT)
    matches = matches[:numGoodMatches]
    # Draw top matches
    imMatches = cv2.drawMatches(im1, keypoints1, im2, keypoints2, matches, None)
    cv2.imwrite("matches.jpg", imMatches)
    # Extract location of good matches
    points1 = np.zeros((len(matches), 2), dtype=np.float32)
    points2 = np.zeros((len(matches), 2), dtype=np.float32)
    
    
    for i, match in enumerate(matches):
        points1[i, :] = keypoints1[match.queryIdx].pt
        points2[i, :] = keypoints2[match.trainIdx].pt
    # Find homography
    h, mask = cv2.findHomography(points1, points2, cv2.RANSAC)
    h = h*np.identity(3)
    h[0,0]=1
    # Use homography
    height, width, channels = im2.shape
    im1Reg = cv2.warpPerspective(im2, h, (width, height))
    return im1Reg, h


# Read reference image
ref = (np.expand_dims(np.repeat(arm_demo_joint_array,repeats=10,axis=1),axis=2)/180*255).astype('uint8')
# Read image to be aligned
base = (np.expand_dims(np.repeat(arm_patient_joint_array,repeats=10,axis=1),axis=2)/180*255).astype('uint8')
# Registered image will be resotred in imReg.
# The estimated homography will be stored in h.
imReg, h = alignImages(base, ref)
print("Estimated homography : \n",  h)
plt.imshow(imReg)
```

```python
def rotate_poses(poses_3d, R, t):
    R_inv = np.linalg.inv(R)
    for pose_id in range(len(poses_3d)):
        pose_3d = poses_3d[pose_id].reshape((-1, 4)).transpose()
        pose_3d[0:3, :] = np.dot(R_inv, pose_3d[0:3, :] - t)
        poses_3d[pose_id] = pose_3d.transpose().reshape(-1)

    return poses_3d

file_path = os.path.join('data', 'extrinsics.json')
with open(file_path, 'r') as f:
    extrinsics = json.load(f)
R = np.array(extrinsics['R'], dtype=np.float32)
t = np.array(extrinsics['t'], dtype=np.float32)
```
